# Support Vector Machine



```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PimaDB=pd.read_csv("https://raw.githubusercontent.com/ettore1461/My-Introduction-to-Statistical-Learning/master/Dataset/PimaDB/PimaDB.csv")
```

```python
PimaDB.head()
```

```python
from sklearn.svm import SVC 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

Importo il **Support Vector Classifier**

```python
sns.stripplot(data=PimaDB,x="Glucose",y="BloodPressure",hue="Outcome")
```

```python
X=PimaDB[["BloodPressure","Glucose"]]
Y=PimaDB.Outcome
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=123)
```

```python
svc=SVC() # Creao il classificatore SVM
```

```python
svc.fit(x_train,y_train)
```

```python
pred=svc.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(pred,y_test)/pred.size,annot=True)
```

```python
print(classification_report(pred,y_test))
```

```python
x1,x2 =np.meshgrid(x_train.iloc[:,0],x_train.iloc[:,1])
```

```python
x1=x1.reshape(np.prod(x1.shape))
x2=x2.reshape(np.prod(x2.shape))
```

```python
pred_grid=svc.predict(pd.DataFrame({"x1":x1,"x2":x2}))
```

```python
sns.stripplot(x=x1,y=x2,hue=pred_grid,alpha=1,size=0.3)
plt.grid()
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train.values[0:10,:],y_train.values[0:10],clf=svc,res=0.02,legend=2)
```

### Approccio con cross-validation

la scelta migliore dei parametri della funzione

```python
parametri={"C":[0.1,1,10,100,1000],"kernel":["rbf"], "gamma":[1,.1,.01,.001,.0001]}
```

```python
from sklearn.model_selection import GridSearchCV # Per care cross-validation sui parametri
```

```python
grid=GridSearchCV(SVC(),parametri,refit=True,verbose=4)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
pd.DataFrame(grid.cv_results_) # raccolgo i risultati del CV
```

```python
pred=grid.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(pred,y_test)/pred.size,annot=True)
```

```python
print(classification_report(pred,y_test))
```
