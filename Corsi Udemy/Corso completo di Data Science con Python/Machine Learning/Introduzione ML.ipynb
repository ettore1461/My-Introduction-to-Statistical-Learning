Con il termine **machine learning** si indica l'insieme di tutti quegli strumenti per l'apprendimento automatico relativo a un problema e il tentativo di dare una previsione sui nuovi dati passati da quel problema

Gli algoritmi di ML possono essere:
* supervisionati
* non supervisionati
* semi-supervisionati
* attivi



I problemi del ML sono:
* overfitting
* underfitting
* troppe variabili

### Regressione
La **correlazione** analizza l'esistenza di una relazione tra le due variabili.

La **regressione** analizza la forma di questa relazione tra le due variabili.
