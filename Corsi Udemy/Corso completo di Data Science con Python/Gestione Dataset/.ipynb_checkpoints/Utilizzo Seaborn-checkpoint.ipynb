# Load Moduli e Dataset

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
Titanic=pd.read_csv("https://raw.githubusercontent.com/ettore1461/My-Introduction-to-Statistical-Learning/master/Dataset/titanic/train.csv")
```

# Analisi 

```python
Titanic.head()
```

```python
sns.countplot(x="Pclass",data=Titanic,palette="rainbow")
```

```python
sns.barplot(x="Pclass",y="Survived",data=Titanic,ci=99.5,palette="rainbow")
```

questo grafico mostra un istogramma, dove sulla X vengono messe le modalità della variabile categoriale, mentre sull'asse Y la media della variabile quantitativa. la barra nera è l'intervallo di confidenza impostato nel grafico.

```python
sns.boxplot(data=Titanic.iloc[:,1:])
```

il box plot è fatto sulle variabili quantitative (numeriche), in questo caso ho escluso l'indice del passeggero

```python
sns.kdeplot(Titanic.Age)
```

Mostra la distribuzione di probabilità marginale della singola variabile aleatoria

```python
sns.violinplot(Titanic.Age,palette="Set1")
```

```python
sns.violinplot(x="Pclass",y="Age",data=Titanic,palette="rainbow")
```

con il **violinsplot** posso vedere le distribuzioni congiunte dell'età per ogni tipo di classe 

```python
sns.stripplot(x="Pclass",y="Age",hue="Survived",data=Titanic
              #,jitter=True
              ,alpha=0.5)
```

```python
sns.stripplot(x="Pclass",y="Age",hue="Survived",data=Titanic
              ,jitter=True
              ,alpha=0.5)
```

lo **stripplot** mostra la densità di punti per ogni classe e età. Da notare che il campo **jitter** distribuisce i punti della stessa coppia (X,Y) in modo orizzontale.

```python
sns.stripplot(x="Pclass",y="Age",hue="Survived",data=Titanic
              ,jitter=True
              ,split=True
              ,alpha=0.5)
```

con il parametro **split** divido le due classi

```python
sns.jointplot(x="Pclass",y="Age",data=Titanic,kind="scatter")
```

```python
sns.jointplot(x="Pclass",y="Age",data=Titanic,kind="hex")
```

la funzione **joinplot** permette di mostrare in grafico bifariato di due variabili del dataset. il campo **kind** permette di modificare la visualizzazione del grafico. con il valore hex posso raggruppare i punti per ottenere una rappresentazione più sintetica di quell'area di punti

```python
sns.lmplot(x="Pclass",y="Age",data=Titanic)
```

```python
sns.lmplot(x="Pclass",y="Age",hue="Survived",data=Titanic)
```

```python
sns.heatmap(Titanic.isnull(),yticklabels=False)
```

Per mostrare visivamente la presanza di valori NAN presenti nel dataset posso usare mostrare una **heatmap** che mostra i valori del metodo **isnull**

```python
sns.set_style("darkgrid")
sns.distplot(Titanic.Age.dropna(),kde=True,color="darkred",bins=30)
```

Per mostrare un istogramma di una variabile, con la funzione di approssimazione **kde**

```python
sns.heatmap(Titanic.corr(),annot=True)
```

Mostro le correlazioni lineari delle variabili quantitative presenti nel dataset 

```python
sns.heatmap(Titanic.isna().iloc[:,11:12])
```

```python
np.sum(Titanic.isna().Embarked)
```

```python
pt1=Titanic.pivot_table(values="Survived",index=["Sex"],columns=["Pclass"],aggfunc=np.mean)
```

```python
sns.heatmap(pt1,cmap="magma")
```

in questo grafico posso mostrare due variabili come se fosse una fotografia

```python
sns.clustermap(pt1,cmap="magma")
```

in questo caso invece vedo un cluster applicato ai dati, andando a raccogliere i dati più simili

```python
sns.pairplot(Titanic.iloc[:,0:5])
```

mostro il **pairplot** del dataset.
